```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create dataframe from csv file:
data = pd.read_csv(r"E:\vaishnav srivastava\stocks\tcs1year.csv", index_col = 'Date')
#dataframe looks like:
data.head()
```

```python
#select 'Close' column
close = data['Close']

#convert date strings in the index into pandas datetime format:

close.index = pd.to_datetime(close.index)

close
```

```python
#calculating Simple moving average by rolling()

sma30 = close.rolling(window=30).mean()

sma30
```

```python
#Calculating 20 days rolling Standard Deviation

std20 = close.rolling(window=20).std()

std20
```

```python
#creating plot

plt.style.use('fivethirtyeight')

plt.figure(figsize = (12,6))

#ploting price and sma :

plt.plot(close, label='TCS Close', linewidth=3)

plt.plot(sma30, label = "30 days SMA", linewidth=1.5)

plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Price with SMA')
plt.legend()

plt.show()
```
